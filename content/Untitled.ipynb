The range of the 32-bit integer is .

When we add several integer values, the resulting sum might exceed the above range. You might need to use long long int in C/C++ or long data type in Java to store such sums.


```python
#!/bin/python3

import sys
def bignumadd(a1,b):
        #a1=list(a)
        b1=list(str(b))
        d=[]

        big = max(len(a1),len(b1))
        small = min(len(a1),len(b1))

        f=0
        for i in range(1, small+1):
            k=int(a1[-i])+int(b1[-i])+f
            f=k/10
            k=k%10
            d.insert(0,k)
            if (big == small and i == big and f > 0):
                d.insert(0,f)
                f = 0
        if big > small :
            for i in range(small+1, big+1):
                k=int(a1[-i])+f
                if (k>10):
                    f=k/10
                    k=k%10
                    d.insert(0,k)
                else:
                    f = 0
                    d.insert(0,k)
        if f > 0:
            d.insert(0,f)
            
        #print (''.join(str(i) for i in d))
        return d
def aVeryBigSum(n, ar):
    # Complete this function
    ret = list(str(ar[0]))
    i = 1 
    while i < n :
        ret = bignumadd(ret,ar[1])
        i += 1
    
    return ret

n = int(input().strip())
ar = list(map(int, input().strip().split(' ')))
result = aVeryBigSum(n, ar)
print(result)
```
